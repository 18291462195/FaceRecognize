```python
#导入模块
from sklearn.svm import SVC

import sklearn.datasets as datasets

from sklearn.model_selection import train_test_split

from sklearn.decomposition import PCA

from sklearn.model_selection import GridSearchCV

import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

import logging
```

使用fetch_lfw_people导入数据，如果本地没有会从网络下载，如果本地有数据，加载本地

```python
faces = datasets.fetch_lfw_people(min_faces_per_person=70,resize = 1)
```

```python
X = faces['data']

y = faces['target']
```

```python
y
```

```python
#测试
image = faces.data[1].reshape(125,94)
plt.imshow(image,cmap = 'gray')
```

分割训练数据和预测数据

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 88)
```

```python
X_train.shape
```

使用PCA进行降维

```python
pca = PCA(n_components=150,whiten=True)

pca.fit(X)
```

```python
X_train_pca = pca.transform(X_train)
X_train_pca.shape
```

```python
X_test_pca = pca.transform(X_test)
```

使用GridSearchCV我们找到了最佳的支持向量机方法

```python
1/150
```

```python
svc = SVC()

c = [0.5,1,3,5]

# 1/n_features = 1/150
gamma = [0.001,0.003,0.006,0.01,0.1]

gcv = GridSearchCV(svc,param_grid={'C':c,'gamma':gamma})
```

```python
gcv.fit(X_train_pca,y_train)
```

```python
svc_best = gcv.best_estimator_
```

```python
gcv.best_score_
```

进行数据预测

```python
y_ = svc_best.predict(X_test_pca)
y_
```

```python
# 数据的可视化
# 11,8列

plt.figure(figsize=(8*2,11*3))

for i in range(88):
    axes = plt.subplot(11,8,i+1)
    
    axes.imshow(X_test[i].reshape(125,94),cmap = 'gray')
    axes.axis('off')
    
#     设置每张图片标题title
#     y_test 真实值 0 ~ 6
#     y_ 预测值0 ~ 6
    t = target_names[y_test[i]].split(' ')[-1]
    p = target_names[y_[i]].split(' ')[-1]
    axes.set_title('True:%s\nPredict:%s'%(t,p))
```
